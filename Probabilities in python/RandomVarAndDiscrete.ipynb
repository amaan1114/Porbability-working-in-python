```python
#Discrete Variables
import numpy as np
from scipy.stats import expon
#Generating Random Variables
var = np.random.choice([1,2,3],1000,p=[0.25,0.35,0.4])

#calculating mean,variance,SD of sample
mean = var.mean()
variance= var.var()
sd=var.std()

print("Mean = ",mean)
print("variance = ",variance)
print("standard deviation = ",sd)
```

```python
#Continuous Random Variable

var2=np.random.exponential(5,2000)
#Visualizing the distribution using histogram
import matplotlib.pyplot as plt
plt.title("2000 random sample from exponential distribution with mean 5\n")
plt.hist(var2,bins=50,alpha=0.6,edgecolor="black",label='Histogram')
plt.show()

#plot pdf
x = np.linspace(0, np.max(var2), 1000)
pdf = expon.pdf(x, scale=5)
plt.plot(x, pdf, 'r',lw=2, label='Theoretical PDF')

```
