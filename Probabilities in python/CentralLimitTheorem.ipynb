```python
import numpy as np
import matplotlib.pyplot as plt
var = np.random.uniform(1,10,10000)
draw = np.random.choice(var,size=(1000,30))
print(draw)
SampleMeans = np.mean(draw) 

#plot uniform distribution
plt.subplot(1, 2, 1)
plt.hist(var, bins=50, color='skyblue', edgecolor='black')
plt.title('Original Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Raw Frequency')

# Plot sample means distribution
plt.subplot(1, 2, 2)
plt.hist(SampleMeans, bins=50, color='lightgreen', edgecolor='black')
plt.title('   | Distribution of Sample Means')
plt.xlabel('Sample Mean')


```
